```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC 
from sklearn.decomposition import PCA
```

# Carga de data

```python
data = pd.read_csv('attrition.csv')
data.head()
```

# Feature Engineering

```python
data.isna().sum()
```

```python
def categorical_cols(df):
    colnames = df.columns
    categorical_vars = []
    for col in colnames:
        if(df[col].dtype == 'object'):
            categorical_vars.append(col)
    return categorical_vars

categorical_cols(data)
```

```python
def labelencodecolumns(df):
  colnames = df.columns
  for col in colnames:
    if(df[col].dtype == 'object'):
      data[col] = data[col].astype('category')
      data[col] = data[col].cat.codes
  return (df.head())

labelencodecolumns(data)
```

## Análisis de data

```python
colnames = data.columns
sn.set(rc={'figure.figsize':(20,50)})
for i, column in enumerate(colnames, 1):
    plt.subplot(11,5,i)
    p=sn.histplot(data[column],color='lightgreen')  
```

```python
col_names = colnames.drop('Attrition', 'EmployeeCount')
col_names = colnames.drop('Over18', 'StandardHours')
sn.set(rc={'figure.figsize':(20,50)})
for i, column in enumerate(col_names[:12], 1):
    x=col_names[i+12]
    plt.subplot(13,2,i)
    p=sn.lineplot(hue='Attrition',x=x,y=column,data=data,color='blue')
```

```python
data.shape
```

### Balanceo de datos

```python
data['Attrition'].value_counts()
```

```python
yes = data[data['Attrition'] == 1]
yes.shape
```

```python
no = data[data['Attrition'] == 0]
no.shape
```

```python
nos = no.sample(2*len(yes), random_state=2022)
nos.shape
```

```python
data = yes.append([nos], ignore_index=True)
data.shape
```

```python
x = data.iloc[:, data.columns != 'Attrition']
y = data['Attrition']
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(x)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, 
                                                    random_state=2022, shuffle=True)
```

## Reducción de dimensionalidad 

### Regresión logística

```python
rlog_clf = LogisticRegression(random_state=0)
rlog_clf.fit(X_train, y_train)
preds_rlog = rlog_clf.predict(X_test)
accuracy_score(y_test, preds_rlog)
```

### SVM

```python
svm_clf = SVC()
svm_clf.fit(X_train, y_train)
preds_svm = svm_clf.predict(X_test)
accuracy_score(y_test, preds_svm)
```

### PCA

```python
pcas = np.arange(2, 34)
accs = []

for pca in pcas:
    #Calculo PCA
    pca = PCA(n_components=pca)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)
    
    #entrenamiento del modelo
    svm_clf_pca = SVC(random_state=0)
    svm_clf_pca.fit(X_train_pca, y_train)
    preds_svm_pca = svm_clf_pca.predict(X_test_pca)
    acc = accuracy_score(y_test, preds_svm_pca)
    accs.append(acc)
    
plt.figure(figsize = (15, 8)),
plt.plot(pcas, accs, "-o", color="red")
plt.title("PCA con SVM")
plt.xlabel("Componentes Principales")
plt.ylabel("ACC")
plt.show()
```

```python
pca = PCA(n_components=22)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
svm_clf_pca = SVC(random_state=0)
svm_clf_pca.fit(X_train_pca, y_train)
preds_svm_pca = svm_clf_pca.predict(X_test_pca)
accuracy_score(y_test, preds_svm_pca)
```

# Arquitectura de red

### Parámentros

```python
parametros = {
    'batch_size': [50, 100],
    'optimizador': ['adam', 'rmsprop'],
    'epochs': [10, 25, 50],
}
```

## Estructura No. 1

```python
def modelo(optimizador):
  Clasificador = Sequential()
  Clasificador.add(Dense(input_dim= X_train.shape[1], units=7, kernel_initializer='uniform', activation='relu'))
  Clasificador.add(Dense(units=14, kernel_initializer='uniform', activation='relu'))
  Clasificador.add(Dense(units=7, kernel_initializer='uniform', activation='relu'))
  Clasificador.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))
  Clasificador.compile(optimizer=optimizador, loss='binary_crossentropy', metrics=['accuracy'])
  return Clasificador
```

```python
resultado = KerasClassifier(modelo, validation_split=0.2, verbose=1)
```

```python
grid_search = GridSearchCV(estimator= resultado, 
                         param_grid= parametros, 
                         scoring='accuracy', 
                         cv=5)
```

```python
grid_search.fit(X_train, y_train, verbose = 1)
```

```python
grid_search.best_params_
```

```python
clasificador1 = Sequential()
clasificador1.add(Dense(input_dim= X_train.shape[1], units=7, kernel_initializer='uniform', activation='relu'))
clasificador1.add(Dense(units=14, kernel_initializer='uniform', activation='relu'))
clasificador1.add(Dense(units=7, kernel_initializer='uniform', activation='relu'))
clasificador1.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))
clasificador1.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
clasificador1.fit(X_train, y_train, batch_size=50, validation_split=0.2, verbose=1, epochs=50)
```

```python
clasificador1.save('7_14_7_rmsprop.h5')
```

```python
model_1 = keras.models.load_model('7_14_7_adam.h5')
```

```python
y_preds = model_1.predict(X_test)
```

```python
y_preds = np.where(y_preds > 0.5, 1, 0)
```

```python
print(classification_report(y_test, y_preds))
```
